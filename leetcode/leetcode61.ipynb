```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def rotateRight(head: ListNode, k: int) -> ListNode:
    if not head or not head.next or k == 0:
        return head

    current = head
    length = 1
    while current.next:
        current = current.next
        length += 1

    tail = current
    tail.next = head

    k = k % length

    steps_to_new_tail = length - k
    for _ in range(steps_to_new_tail):
        tail = tail.next

    newHead = tail.next
    tail.next = None

    return newHead
```
