```python
def climbStairs(n: int) -> int:
    if n == 1:
        return 1
    
    if n == 2:
        return 2

    prev_step = 1
    current_step = 2
    
    for _ in range(3, n + 1):
        next_step = prev_step + current_step
        prev_step = current_step
        current_step = next_step
        
    return current_step
```
