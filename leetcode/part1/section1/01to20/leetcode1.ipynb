```python
def twoSum(nums, target):
    num_map = {}

    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        
        num_map[num] = i
    
    return []
```

## ðŸ’¡ The "Two Sum" Problem: A LeetCode Classic

The "Two Sum" problem, officially LeetCode problem number 1, is arguably the most famous and foundational question in competitive programming and technical interviews. It serves as an excellent entry point for beginners to understand basic algorithm design, specifically focusing on how to optimize time complexity, and it beautifully illustrates the power of using appropriate data structures like hash maps (or hash tables).

---

### ðŸ“ Problem Statement and Goal

The problem asks you to find two distinct indices, $i$ and $j$, within a given array of integers, $nums$, such that the values at these indices, $nums[i]$ and $nums[j]$, add up exactly to a specific integer target. You are guaranteed that exactly one solution exists, and you may not use the same element twice. The ultimate goal is to return an array or list containing these two indices, $[i, j]$. For instance, if $nums = [2, 7, 11, 15]$ and $target = 9$, the required output is $[0, 1]$ because $nums[0] + nums[1] = 2 + 7 = 9$.

---

### âš™ï¸ Brute-Force Solution (Inefficient)

The most straightforward way to solve this is through a **brute-force** approach. This involves using a nested loop structure. The outer loop iterates through every element $nums[i]$, and the inner loop checks every *subsequent* element $nums[j]$ (where $j > i$) to see if their sum equals the $target$. If $nums[i] + nums[j] = target$, you immediately return $[i, j]$. While correct, this method forces every possible pair to be checked. If the array has $N$ elements, the number of pairs checked is proportional to $N^2$, giving this approach a time complexity of $O(N^2)$, which is generally too slow for large inputs.

---

### ðŸ§  Optimal Solution: Using a Hash Map (Efficient)

The most efficient and widely accepted solution leverages a **hash map** (also known as a dictionary or hash table) to reduce the time complexity from quadratic to linear. The key insight is to realize that for any number $x$ in the array, the required **complement** number needed to reach the target is $y = target - x$.  Instead of searching the *rest* of the array for $y$ (which takes $O(N)$ time in the brute-force approach), we can pre-store or simultaneously store the array elements and their indices in the hash map.

---

### ðŸ’» The Hash Map Walkthrough

The algorithm performs a single pass through the $nums$ array. For each element $nums[i]$: first, it calculates the **complement** needed: $complement = target - nums[i]$. Next, it checks if this $complement$ already exists as a **key** in the hash map. If it does, it means we have found the two numbers, and we return the current index $i$ along with the index associated with the $complement$ (which is the value stored in the map). If the $complement$ is *not* found in the map, the algorithm adds the current number $nums[i]$ and its index $i$ to the hash map, using $nums[i]$ as the key and $i$ as the value. Because looking up or inserting an element in a hash map takes, on average, $O(1)$ time, this single-pass approach achieves a much faster **linear time complexity of $O(N)$**.
