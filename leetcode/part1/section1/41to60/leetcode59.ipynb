```python
def generateMatrix(n: int) -> list[list[int]]:
    
    matrix = [[0] * n for _ in range(n)]
    
    top, bottom, left, right = 0, n - 1, 0, n - 1
    current = 1
    
    while top <= bottom and left <= right:
        
        for c in range(left, right + 1):
            matrix[top][c] = current
            current += 1
        top += 1
        
        for r in range(top, bottom + 1):
            matrix[r][right] = current
            current += 1
        right -= 1
        
        if top <= bottom:
            for c in range(right, left - 1, -1):
                matrix[bottom][c] = current
                current += 1
            bottom -= 1
            
        if left <= right:
            for r in range(bottom, top - 1, -1):
                matrix[r][left] = current
                current += 1
            left += 1
            
    return matrix
```

## ðŸ—ï¸ LeetCode Problem 59: Spiral Matrix II Explained

LeetCode problem 59, "Spiral Matrix II," is a matrix construction problem that requires generating an $N \times N$ square matrix and filling it with integers from $1$ to $N^2$ in a **spiral order**. The output must be the complete $N \times N$ matrix. This problem is the inverse of "Spiral Matrix" (Problem 54), where the task was to traverse and return the elements of an existing matrix in spiral order. Here, we build the matrix while following the spiral path.

---

The most effective method to solve this is a **boundary-tracking simulation** approach, similar to the one used in Spiral Matrix I. We use four variables to define the boundaries of the current square layer we are filling: `top`, `bottom`, `left`, and `right`. Since the matrix is $N \times N$, initially, $\text{top} = 0$, $\text{bottom} = N-1$, $\text{left} = 0$, and $\text{right} = N-1$. We also need a counter variable, $\text{current\_number}$, initialized to $1$, to fill the matrix cells sequentially up to $N^2$.

---

The algorithm operates within a `while` loop that continues as long as $\text{top} \le \text{bottom}$ and $\text{left} \le \text{right}$. In each iteration of the loop, we fill one full **layer** of the spiral in four distinct movements: **Right, Down, Left, Up**. After each direction's traversal, we adjust the corresponding boundary inwards, effectively shrinking the area for the next iteration. 

---

**Step 1: Traverse Right (Top Row):** We iterate across the current `top` row from `left` column to `right` column. For each column $j$, we set $\text{matrix}[\text{top}][j] = \text{current\_number}$ and then increment $\text{current\_number}$. After filling this row, we move the top boundary inwards by incrementing $\text{top}$ ($\text{top} = \text{top} + 1$).

---

**Step 2: Traverse Down (Right Column):** We then traverse the current `right` column from the new `top` row down to the `bottom` row. For each row $i$, we set $\text{matrix}[i][\text{right}] = \text{current\_number}$ and increment the counter. Once this vertical segment is filled, we move the right boundary inwards by decrementing $\text{right}$ ($\text{right} = \text{right} - 1$).

---

**Step 3: Traverse Left (Bottom Row):** Next, we check if $\text{top} \le \text{bottom}$ (a crucial check to handle the center element of odd-$N$ matrices). If the condition holds, we traverse the current `bottom` row backward, from the new `right` column to the `left` column. For each column $j$, we set $\text{matrix}[\text{bottom}][j] = \text{current\_number}$ and increment the counter. We then move the bottom boundary inwards by decrementing $\text{bottom}$ ($\text{bottom} = \text{bottom} - 1$).

---

**Step 4: Traverse Up (Left Column):** Finally, we check if $\text{left} \le \text{right}$ (another crucial check). If the condition holds, we traverse the current `left` column upward, from the new `bottom` row up to the new `top` row. For each row $i$, we set $\text{matrix}[i][\text{left}] = \text{current\_number}$ and increment the counter. We then complete the layer by moving the left boundary inwards by incrementing $\text{left}$ ($\text{left} = \text{left} + 1$).

---

The loop repeats, filling successively smaller inner layers until the boundaries cross ($\text{top} > \text{bottom}$ or $\text{left} > \text{right}$). The boundary checks in steps 3 and 4 are necessary to prevent over-writing the single central element when $N$ is odd, which would have already been filled in one of the previous steps. Once the loop terminates, the $N \times N$ matrix is fully populated with numbers from $1$ to $N^2$ in the correct spiral order. The time complexity is $O(N^2)$, as we visit and fill each of the $N^2$ cells exactly once. The space complexity is also $O(N^2)$ to store the resulting matrix.
