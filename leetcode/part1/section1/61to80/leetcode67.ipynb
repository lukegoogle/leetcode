```python
def addBinary(a: str, b: str) -> str:
    res = []
    i, j = len(a) - 1, len(b) - 1
    carry = 0

    while i >= 0 or j >= 0 or carry:
        total_sum = carry

        if i >= 0:
            total_sum += int(a[i])
            i -= 1
        if j >= 0:
            total_sum += int(b[j])
            j -= 1

        res.append(str(total_sum % 2))
        carry = total_sum // 2

    return "".join(res[::-1])
```
