```python
def fullJustify(words: list[str], maxWidth: int) -> list[str]:
    res = []
    n = len(words)
    i = 0

    while i < n:
        line_words = []
        current_len = 0
        j = i
        while j < n and current_len + len(words[j]) + len(line_words) <= maxWidth:
            current_len += len(words[j])
            line_words.append(words[j])
            j += 1
        
        num_words = len(line_words)
        total_spaces = maxWidth - current_len
        
        is_last_line = j == n
        
        if num_words == 1 or is_last_line:
            line = " ".join(line_words)
            line += " " * (maxWidth - len(line))
        else:
            gaps = num_words - 1
            base_spaces = total_spaces // gaps
            extra_spaces = total_spaces % gaps
            
            line = ""
            for k in range(num_words):
                line += line_words[k]
                if k < gaps:
                    num_spaces = base_spaces
                    if k < extra_spaces:
                        num_spaces += 1
                    line += " " * num_spaces
        
        res.append(line)
        i = j
        
    return res
```
