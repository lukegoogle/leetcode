The LeetCode problem **Two Sum** is frequently the very first problem encountered by aspiring programmers on the platform, serving as a foundational test of array manipulation and basic data structure knowledge. The challenge is deceptively simple: given an array of integers, $nums$, and an integer target, return the indices of the two numbers in $nums$ that add up to the target. The problem explicitly states that there will be exactly one valid solution, and you may not use the same element twice. This constraint means we are searching for two distinct positions, $i$ and $j$, such that $nums[i] + nums[j] = target$.

***

The most straightforward method to solve this, the **Brute-Force Approach**, is to check every single combination of two numbers in the array. This involves using two nested loops. The outer loop iterates with index $i$ from the beginning of the array up to the second-to-last element. The inner loop, controlled by index $j$, starts from $i + 1$ (to avoid using the same element twice and to prevent checking pairs already evaluated) and goes to the end of the array. Inside the inner loop, a simple conditional check is performed: if $nums[i] + nums[j]$ equals $target$, the indices $i$ and $j$ are returned immediately. This method is easy to understand and implement but is highly inefficient.

***

The inefficiency of the brute-force method stems directly from its **time complexity**. If the array contains $N$ elements, the outer loop runs $N-1$ times, and the inner loop runs approximately $N$ times in the worst-case scenario (when the solution is the last pair). This results in a total number of operations proportional to $N^2$. Therefore, the time complexity is $O(N^2)$. For small arrays, this is acceptable, but for large arrays with thousands or tens of thousands of elements, an $O(N^2)$ solution quickly becomes too slow, causing a **Time Limit Exceeded (TLE)** error on the LeetCode judge. A more performant solution is required to satisfy the time constraints of typical programming interviews and competitions.

***

To dramatically improve the performance, we must find a way to avoid the inner loop. The key insight is to realize that for any given number $x$ (which is $nums[i]$), the second number required to hit the target, let's call it the **complement**, must be $target - x$. If we could instantly know whether this complement exists in the remaining part of the array, we could solve the problem in a single pass. This realization naturally leads us to consider using a data structure optimized for fast lookups.

***

The optimal data structure for solving the Two Sum problem is the **Hash Map**, also known as a dictionary or hash table. Hash maps are designed to provide near-instantaneous average time complexity for both insertion and retrieval operations, specifically $O(1)$. By using a hash map, we can effectively trade some space complexity for a massive gain in time complexity, a common and valuable technique in algorithm design.

***

The **One-Pass Hash Map Approach** works by iterating through the array only once. As we process each number $nums[i]$ at index $i$, we first calculate its required complement: $complement = target - nums[i]$. Instead of looking forward in the rest of the array, we check our hash map to see if we have **already processed** this complement in a previous iteration.

***

If the calculated complement is found as a key in the hash map, it means we have previously encountered the first number of our pair. Since the hash map stores the index of that previous number as its value, we can immediately return the index stored in the map for the complement and the current index $i$. This is the moment the solution is found, and the function terminates.

***

If the complement is **not** found in the hash map, it means the current number, $nums[i]$, is not the second number of any pair we have previously started forming. In this case, the current number itself might be the first number of a future pair. We therefore add $nums[i]$ as a key to the hash map, setting its value to its current index $i$, and then proceed to the next element in the array. This ensures that any subsequent elements can instantly check against the number we just processed.

***

The major advantage of the hash map approach is its **linear time complexity**, $O(N)$. Because the entire list is traversed only once, and each lookup or insertion operation takes an average of $O(1)$ time, the total execution time scales linearly with the size of the input array. This $O(N)$ solution is highly efficient and is the intended, accepted solution for this problem, comfortably passing the time limit constraints.

***

In summary, the transition from the $O(N^2)$ brute-force solution to the $O(N)$ hash map solution illustrates a fundamental principle in algorithm optimization: leveraging an appropriate data structure (the hash map, in this case) to convert a slow search operation within a loop into a fast, constant-time lookup operation.  This simple, yet powerful, technique is essential for solving many other related algorithmic problems.

```python
# leetcode 1

def twoSum(nums, target):
    num_map = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        num_map[num] = i

twoSum([0,1,2,3,4,5,6,7],10)
```

The LeetCode problem **Add Two Numbers** (Problem 2) is a foundational exercise in handling and manipulating the **Linked List** data structure. The core challenge involves taking two separate linked lists, where each list represents a non-negative integer, and computing their sum, which must also be returned as a new linked list. A critical constraint to remember is that the digits are stored in **reverse order**, meaning the head of the list holds the least significant digit (the "ones" place).

***

The standard way humans perform multi-digit addition involves aligning the numbers vertically and summing them column by column, starting from the rightmost (least significant) digit. If a column sum exceeds nine, a **carry** (a 1) is generated and added to the sum of the next column to the left. The structure of the input in this problem perfectly mimics this manual process because the linked lists are already aligned for addition: the head node of $l1$ corresponds directly to the head node of $l2$ (the ones place), the second node corresponds to the tens place, and so on.

***

The most efficient solution, often called the **One-Pass Iterative Approach**, directly translates this manual addition process into an algorithm. Since we are moving through the lists sequentially, from head to tail, we are naturally processing the digits from least significant to most significant. This single pass eliminates the need for any complex reversal or conversion of the linked lists to standard integers, which could cause **integer overflow** issues if the numbers were extremely large.

***

To manage the construction of the new result list, the algorithm begins by creating a special node: the **dummy head**. This node does not store any part of the final sum but acts as a convenient placeholder. It simplifies the code by providing a fixed starting point and handling the otherwise tricky logic of initializing the first actual node of the result list. A `current` pointer is initialized to this dummy head, and all subsequent new sum nodes will be attached to `current.next`.

***

Another crucial variable initialized at the start is the **carry**, set to zero. The loop then begins, and it must continue running as long as there are digits remaining in *either* of the input lists ($l1$ or $l2$) **OR** if the `carry` variable holds a value greater than zero. The inclusion of the carry condition is vital, as the final sum might require one extra digit (e.g., $99 + 1 = 100$, requiring a final carry node).

***

Inside the loop, for the current position, the algorithm retrieves the values from $l1$ and $l2$. If either list has already been exhausted (i.e., its pointer is `None`), its value is treated as $0$. The core calculation involves summing three components: $\text{val}_1$ (from $l1$), $\text{val}_2$ (from $l2$), and the $\text{carry}$ from the previous step. 

***

Once the **total sum** ($S$) for the current column is calculated, it is used to determine two things: the value of the new node and the carry for the next column. The value of the new node is the digit in the current place, calculated using the modulo operator: $\text{digit} = S \pmod{10}$. The carry value for the next iteration is the quotient from integer division: $\text{carry} = S // 10$.

***

A new `ListNode` is created using this calculated $\text{digit}$ and appended to the result list by setting `current.next` to the new node. The `current` pointer is then advanced to this newly created node (`current = current.next`). After this, the pointers for $l1$ and $l2$ are advanced to their respective next nodes only if they are not yet `None`.

***

The loop continues until all three termination conditions are met: $l1$ is exhausted, $l2$ is exhausted, and the `carry` is zero. Upon termination, the complete sum list is formed. The final result returned by the function is not the `dummy\_head` itself, but the node immediately following it: `dummy\_head.next`. This cleanly excises the initial placeholder node, leaving only the required sum list.

***

This iterative approach has an optimal **Time Complexity** of $O(\max(m, n))$, where $m$ and $n$ are the lengths of the two input lists. Since the algorithm traverses each list only once, the time taken scales linearly with the length of the longer list. Similarly, the **Space Complexity** is also $O(\max(m, n))$, as the space required is directly proportional to the length of the new linked list generated to store the sum. This combination of speed and efficient space usage makes the single-pass iterative solution the standard, accepted answer for LeetCode Problem 2.

```python
# leetcode 2

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1: ListNode, l2: ListNode) -> ListNode:
    dummy_head = ListNode(0)
    current = dummy_head
    carry = 0

    while l1 or l2 or carry:
        val1 = l1.val if l1 else 0
        val2 = l2.val if l2 else 0

        total_sum = val1 + val2 + carry
        
        carry = total_sum // 10
        new_digit = total_sum % 10

        current.next = ListNode(new_digit)
        current = current.next

        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next

    return dummy_head.next
```

### ðŸŽ¯ The Problem: Longest Substring Without Repeating Characters (LeetCode 3)

LeetCode Problem 3, titled "Longest Substring Without Repeating Characters," is a fundamental challenge in string manipulation. The goal is to take a given string, $s$, and find the length of the **longest substring** within it that contains **no repeating characters**. It is crucial to note the difference between a *substring* (a contiguous sequence of characters) and a *subsequence* (characters that appear in order, but not necessarily contiguous).

For example, given the input string $s = \text{"abcabcbb"}$, the longest substring without repeating characters is $\text{"abc"}$, and the correct output is its length, 3. For $s = \text{"pwwkew"}$, the longest substring is $\text{"wke"}$, also with a length of 3 (note that $\text{"pwke"}$ is a subsequence, not a substring).

---

### ðŸ’¡ The Efficient Approach: Sliding Window

The most efficient and widely accepted solution for this problem is the **Sliding Window** technique. This method allows us to solve the problem in $O(n)$ time complexity, where $n$ is the length of the string, which is a significant improvement over the brute-force $O(n^3)$ approach. A sliding window maintains a dynamic range (a "window") within the string, represented by two pointers, typically a left pointer ($l$) and a right pointer ($r$). 

The core idea is to expand the window from the right ($r$) as long as the window contains only unique characters. When a repeating character is encountered, we contract the window from the left ($l$) just enough to remove the duplicate and restore the uniqueness property.

---

### ðŸ› ï¸ Data Structures and Algorithm Steps

To implement the sliding window effectively, we need a way to efficiently track the characters currently inside the window and check for duplicates. A **Set** (or a Hash Map/Dictionary) is the ideal data structure for this, as it allows for $O(1)$ average time complexity for adding, checking for existence, and removing characters.

1.  **Initialization:** Initialize two pointers, $l$ (left) and $r$ (right), to 0. Initialize a variable $\text{maxLength}$ to 0 to store the result, and an empty $\text{Set}$ (or Hash Map) called $\text{charSet}$ to store the characters within the current window $[l, r]$.

2.  **Expansion and Validation (The Loop):** Iterate through the string with the right pointer ($r$) from the beginning to the end. In each iteration, we try to include the character at $s[r]$ into the window.

3.  **Handling Duplicates (Contraction):** Before adding $s[r]$, check if it is already present in $\text{charSet}$. If $s[r]$ is already in $\text{charSet}$ (meaning a duplicate is found), we enter a **while loop** to shrink the window from the left. Inside this loop, we remove the character at $s[l]$ from $\text{charSet}$ and increment $l$ by 1. This shrinking continues until the duplicate character at $s[r]$ is no longer present inside the window (i.e., it has been completely removed from the left side).

4.  **Adding and Updating:** Once the window is valid (no duplicates), we add the new character $s[r]$ to $\text{charSet}$. The current length of the valid substring is $r - l + 1$. We then update $\text{maxLength}$ with the maximum of its current value and the new window length.

5.  **Final Result:** After the right pointer $r$ has traversed the entire string, $\text{maxLength}$ holds the length of the longest substring without repeating characters, which is then returned.

---

### â±ï¸ Complexity Analysis

The sliding window approach is highly efficient:

* **Time Complexity: $O(n)$**
    Each character in the string is processed by the right pointer $r$ exactly once. Crucially, each character is also removed by the left pointer $l$ at most once. Since the operations on the $\text{Set}$ (insert, delete, check existence) are all $O(1)$ on average, the total time complexity is linear, proportional to the length of the string, $n$.

* **Space Complexity: $O(k)$**
    The space complexity is determined by the $\text{Set}$ used to store characters. In the worst case, the set will hold all unique characters in the string. If the character set is a fixed size (e.g., 26 for English letters or 128 for ASCII characters), the space complexity is $O(1)$. If the size of the character set, $k$, is unbounded or large (like Unicode), the space complexity is $O(k)$ or $O(\min(n, | \Sigma |))$, where $|\Sigma|$ is the size of the character set.

```python
# leetcode 3

def lengthOfLongestSubstring(s: str) -> int:
    char_set = set()
    l = 0
    max_length = 0
    
    for r in range(len(s)):
        while s[r] in char_set:
            char_set.remove(s[l])
            l += 1
        
        char_set.add(s[r])
        max_length = max(max_length, r - l + 1)
        
    return max_length
```

```python
# leetcode 4

def findMedianSortedArrays(nums1: list[int], nums2: list[int]) -> float:
    m = len(nums1)
    n = len(nums2)

    if m > n:
        nums1, nums2, m, n = nums2, nums1, n, m

    low = 0
    high = m
    halfLen = (m + n + 1) // 2

    while low <= high:
        i = (low + high) // 2
        j = halfLen - i

        if i < m and nums2[j-1] > nums1[i]:
            low = i + 1
        elif i > 0 and nums1[i-1] > nums2[j]:
            high = i - 1
        else:
            maxLeft = 0
            if i == 0:
                maxLeft = nums2[j-1]
            elif j == 0:
                maxLeft = nums1[i-1]
            else:
                maxLeft = max(nums1[i-1], nums2[j-1])

            if (m + n) % 2 == 1:
                return maxLeft

            minRight = 0
            if i == m:
                minRight = nums2[j]
            elif j == n:
                minRight = nums1[i]
            else:
                minRight = min(nums1[i], nums2[j])

            return (maxLeft + minRight) / 2.0
```
