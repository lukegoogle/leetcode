```python
def insert(intervals: list[list[int]], newInterval: list[int]) -> list[list[int]]:
    
    result = []
    
    for i in range(len(intervals)):
        
        if newInterval[1] < intervals[i][0]:
            result.append(newInterval)
            return result + intervals[i:]
        
        elif newInterval[0] > intervals[i][1]:
            result.append(intervals[i])
            
        else:
            newInterval[0] = min(newInterval[0], intervals[i][0])
            newInterval[1] = max(newInterval[1], intervals[i][1])
            
    result.append(newInterval)
    return result
```

## üìê LeetCode Problem 57: Insert Interval Explained

LeetCode problem 57, "Insert Interval," requires you to take an existing sorted list of non-overlapping intervals and a new, single interval, and insert this new interval into the list such that the resulting list remains sorted and non-overlapping. If inserting the new interval causes overlaps with one or more existing intervals, these overlapping segments must be merged into a single, combined interval. This problem is an extension of the "Merge Intervals" concept (Problem 56).

---

The input typically consists of an array of intervals, $\text{intervals}$, which is already sorted by start time and contains no overlaps, and a single $\text{newInterval} = [\text{new\_start}, \text{new\_end}]$. The key to an efficient $O(N)$ solution, where $N$ is the number of existing intervals, is to process the list in three logical stages: **Before Overlap**, **Overlapping and Merging**, and **After Overlap**. Since the initial array is sorted, we can use a single pass to handle all three stages.

---

**Stage 1: Intervals Before Overlap (Pre-Merge):** We iterate through the existing $\text{intervals}$ list. Any interval $[\text{start}, \text{end}]$ that comes entirely before the $\text{newInterval}$ is guaranteed not to overlap. This condition is met if $\text{end} < \text{new\_start}$. All such intervals are simply added directly to the result list in the order they appear. We stop this stage as soon as we encounter an interval that either overlaps with or comes after the $\text{newInterval}$. 

---

**Stage 2: Overlapping and Merging:** This stage begins when we encounter the first interval that overlaps with the $\text{newInterval}$. An overlap occurs if the existing interval's start time is less than or equal to the $\text{newInterval}$'s end time, and the existing interval's end time is greater than or equal to the $\text{newInterval}$'s start time. Since we've passed Stage 1, the condition simplifies to $\text{current\_start} \le \text{new\_end}$. As long as this overlap condition holds, we perform a merge operation.

---

The merge operation does not immediately add an interval to the result list; instead, it modifies the $\text{newInterval}$ itself. The new, combined start time becomes $\min(\text{new\_start}, \text{current\_start})$, and the new combined end time becomes $\max(\text{new\_end}, \text{current\_end})$. We update the $\text{newInterval}$ with these consolidated bounds and continue checking the next interval in the input list. This process effectively absorbs all overlapping intervals into one single, continuously growing $\text{newInterval}$.

---

The merging loop terminates when we find the first existing interval that does **not** overlap with the updated $\text{newInterval}$. This non-overlap condition is met when the existing interval's start time is greater than the $\text{newInterval}$'s end time, i.e., $\text{current\_start} > \text{new\_end}$. Once the merging loop is finished, the consolidated $\text{newInterval}$ is finalized and is added to the result list. 

---

**Stage 3: Intervals After Overlap (Post-Merge):** After adding the finalized, merged interval to the result list, the remaining intervals in the input list are guaranteed to be non-overlapping with it (and thus, non-overlapping with each other as well). This is because the input was initially sorted, and all overlaps were handled in Stage 2. Therefore, all the remaining intervals are simply appended to the result list one by one.

---

The algorithm finishes by returning the completed result list. The efficiency of this three-stage, single-pass approach is $O(N)$ time complexity, as we only iterate through the $N$ existing intervals a single time. The space complexity is $O(N)$ to store the output list. The logic is sound because the pre-sorted nature of the input allows us to determine the boundaries of the merge area purely by checking the start and end times in sequential order.
