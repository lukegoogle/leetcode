```python
def convert(s: str, numRows: int) -> str:
    if numRows == 1 or len(s) <= numRows:
        return s

    rows = [''] * numRows
    current_row = 0
    going_down = False
    
    for char in s:
        rows[current_row] += char
        
        # Change direction at the top and bottom rows
        if current_row == 0 or current_row == numRows - 1:
            going_down = not going_down
            
        # Update row index based on direction
        if going_down:
            current_row += 1
        else:
            current_row -= 1
            
    # Join the rows to form the final string
    return "".join(rows)
```

## ðŸ§µ LeetCode Problem 6: Zigzag Conversion

LeetCode problem 6, "Zigzag Conversion," is a string manipulation problem that is visually oriented and tests a programmer's ability to translate a complex spatial pattern into a sequential, linear algorithm. It involves converting a given string into a zigzag pattern across a specified number of rows, and then reading the resulting string line by line.

---

### ðŸŽ¨ The Zigzag Pattern

The input string, $S$, is written in a zigzag (or "Z" formation) on a fixed number of rows, $numRows$. The characters are placed sequentially, moving vertically down until the last row, then diagonally up and right until the first row, and repeating this cycle. For example, with $numRows = 3$, the pattern looks like:
$$
\begin{array}{ccccccc}
P & & A & & H & & N \\
& A & & P & & L & \\
& & Y & & I & & R
\end{array}
$$
The output is formed by reading the characters row by row, from left to right. For the example above, the output string would be "PAHNAPLSIIGYIR" (the missing characters are from the input, which was "PAYPALISHIRING").

---

### ðŸ“ Determining the Cycle Length

The key to solving this problem efficiently is recognizing the underlying pattern and calculating the **cycle length**. A full cycle (downward stroke + upward diagonal stroke) involves a specific number of steps. In a configuration with $numRows$ rows, the downward stroke uses $numRows$ characters, and the upward stroke uses $numRows - 2$ characters. Therefore, the total number of characters in a complete zigzag cycle is $CycleLength = 2 \times numRows - 2$.  This calculation is the foundation for determining the indices of the characters that belong to the same row.

---

### ðŸ’» Row-by-Row Construction

The optimal $O(N)$ solution involves initializing an empty string (or array of strings/string builders) for each of the $numRows$. The algorithm then iterates through the original string $S$ once. Instead of trying to simulate the movement, we calculate which row each character $S[i]$ belongs to. The current row index can be determined by tracking the current direction (down or up) and the current row number.

---

### ðŸ§  The Index Mapping Approach

A more direct and often simpler $O(N)$ approach avoids simulating the movement entirely. We iterate through each row, from $r=0$ to $r=numRows-1$. For a given row $r$, we know that the first character will be at index $r$. Subsequent characters in the same row will appear after a certain fixed offset.
* **For the top row ($r=0$) and the bottom row ($r=numRows-1$):** The characters are spaced exactly $CycleLength$ apart.
* **For the intermediate rows ($0 < r < numRows - 1$):** Characters appear in pairs with two different offsets. The first offset is $CycleLength$ minus the offset to the second character (which is $2 \times r$). The second offset is $2 \times r$. The algorithm iterates starting from index $r$, and continuously jumps by these two alternating offsets until it exceeds the length of $S$. By appending all these characters in order, the final output string is constructed row by row.
