```python
def lengthOfLastWord(s: str) -> int:
    
    s = s.strip()
    
    if not s:
        return 0
        
    last_space_index = s.rfind(' ')
    
    return len(s) - 1 - last_space_index
```

## üìè LeetCode Problem 58: Length of Last Word Explained

LeetCode problem 58, "Length of Last Word," is a string manipulation task that requires you to determine the length of the last word in a given string $s$. A "word" is defined as a maximal substring consisting only of non-space characters. The main complexities of this problem arise from handling various edge cases, particularly leading, trailing, and multiple consecutive spaces within the input string.

---

The most straightforward and often language-dependent approach involves utilizing built-in string methods. Many programming languages provide functions to $\text{trim}$ leading and trailing whitespace from a string and then $\text{split}$ the string into an array of words based on the space delimiter. Once the array of words is obtained, the length of the last element in that array gives the desired result. While elegant and concise, this method can sometimes be less performant than a custom implementation, especially when dealing with very long strings or memory constraints, as it often creates intermediate string and array objects.

---

The more robust and performance-oriented solution is to implement a **single-pass, two-pointer, or reverse iteration** approach that handles the whitespace manually. This method avoids the overhead of creating new strings or arrays and offers $O(N)$ time complexity, where $N$ is the length of the string, and $O(1)$ extra space complexity. The core idea is to start from the end of the string and work backwards.

---

**Step 1: Skip Trailing Spaces.** We initialize a pointer, say $i$, to the last index of the string ($N-1$). We then move this pointer backward as long as the character at $s[i]$ is a space. This effectively removes any trailing whitespace. If the pointer moves past the beginning of the string during this step, it means the entire string was either empty or consisted only of spaces, in which case the length of the last word is $0$.

---

**Step 2: Measure the Last Word.** Once the pointer $i$ stops at the last non-space character, it marks the end of the last word. We then start a counter, $\text{length}$, initialized to $0$. We continue moving the pointer $i$ backward, incrementing $\text{length}$ for every non-space character encountered. This loop continues until we either hit the beginning of the string (index $i < 0$) or encounter another space character. ---

**Step 3: Return the Length.** When the loop in Step 2 terminates, the value of $\text{length}$ represents the count of non-space characters from the end of the string up to the first space or the string's start. This value is precisely the length of the last word. This manual iteration method is highly efficient as it performs all necessary work in a single pass from right to left, requiring no auxiliary data structures.
