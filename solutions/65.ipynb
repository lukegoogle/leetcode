```python
def isNumber(s: str) -> bool:
    state = 0
    transitions = [
        {'+': 1, '-': 1, '.': 3, 'd': 2, 'e': 8},
        {'.': 4, 'd': 2, 'e': 8},
        {'.': 4, 'd': 2, 'e': 5},
        {'d': 4, 'e': 8},
        {'.': 8, 'd': 4, 'e': 5},
        {'+': 6, '-': 6, 'd': 7, 'e': 8, '.': 8},
        {'d': 7, 'e': 8, '.': 8},
        {'d': 7, 'e': 8, '.': 8},
        {}
    ]

    for char in s:
        char_type = 'e'
        if char.isdigit():
            char_type = 'd'
        elif char in ['+', '-']:
            char_type = char
        elif char == '.':
            char_type = '.'

        if char_type not in transitions[state]:
            return False

        next_state = transitions[state][char_type]
        if next_state == 8:
            return False
        state = next_state

    return state in [2, 4, 7]
```
