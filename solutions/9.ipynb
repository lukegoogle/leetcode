```python
def isPalindrome(x: int) -> bool:
    if x < 0 or (x % 10 == 0 and x != 0):
        return False

    reverted_number = 0
    while x > reverted_number:
        reverted_number = reverted_number * 10 + x % 10
        x //= 10
        
    return x == reverted_number or x == reverted_number // 10
```

## ğŸ”„ LeetCode Problem 9: Palindrome Number - In Immense Detail

LeetCode problem 9, "Palindrome Number," asks for a determination of whether a given integer $x$ is a palindrome. A number is a palindrome if it reads the same backward as forward (e.g., 121, 55, 1001). This problem, while seemingly simple, tests the understanding of integer manipulation, particularly how to reverse a number without converting it to a string, and how to prevent potential integer overflow during the reversal process.

---

### ğŸ§ Fundamental Definition and Edge Cases

The fundamental requirement is a comparison between the original number and its reversed counterpart. However, several critical edge cases must be addressed immediately to ensure a correct and efficient solution:
1.  **Negative Numbers:** Any negative number cannot be a palindrome. For instance, $-121$ reversed is $121-$, which is not $-121$. Thus, if $x < 0$, the function should immediately return `false`.
2.  **Numbers Ending in Zero:** Any number that ends in $0$ (and is not $0$ itself) cannot be a palindrome, as the reversed number would start with $0$, which is impossible for an integer (e.g., $120$ reversed is $021$). Thus, if $x \ne 0$ and $x \pmod{10} = 0$, the function should return `false`. This step also correctly handles the $x=0$ case, which is a palindrome.

---

### ğŸ”¢ The Optimized Numerical Reversal Technique

The most efficient solution avoids string conversion entirely, which saves space and potentially time. Instead, it reverses only **half** of the integer. This is achieved by iteratively building a `reversed_half` number from the least significant digits of the original number $x$, using the modulo and division operators. The loop continues as long as the original number $x$ is greater than the `reversed_half` number being constructed.

In each iteration:
1.  The last digit of $x$ is extracted: $pop = x \pmod{10}$.
2.  The `reversed_half` is updated: $reversed\_half = reversed\_half \times 10 + pop$.
3.  The original $x$ is truncated: $x = \lfloor x / 10 \rfloor$.

---

### âœ‚ï¸ Termination and Comparison: Halving the Work

The loop terminates when $x$ is no longer greater than `reversed_half`. This is the clever optimization: by the time the loop breaks, $x$ will hold the first half of the digits, and `reversed_half` will hold the reversed second half.

There are two scenarios for the final check:
1.  **Even Number of Digits:** If the original number had an even number of digits (e.g., $1221$), the loop terminates when $x$ equals `reversed_half` (e.g., $x=12$ and $reversed\_half=12$). The result is `true`.
2.  **Odd Number of Digits:** If the original number had an odd number of digits (e.g., $121$), the loop terminates when $x$ is equal to the `reversed_half` divided by 10 (e.g., $x=1$ and $reversed\_half=21$). The middle digit (in this case, $2$) is irrelevant for the palindrome check, so it's dropped from `reversed_half`. The result is `true` if $x = \lfloor reversed\_half / 10 \rfloor$. 

---

### â±ï¸ Complexity and Overflow Prevention

This half-reversal method ensures the time complexity remains $O(\log_{10} x)$, or proportional to the number of digits $D$, as only $D/2$ iterations are performed. Crucially, by stopping the reversal halfway, the algorithm inherently prevents the **integer overflow** that was the primary complication in LeetCode 7. Since we only reverse half the digits, the `reversed_half` number will never exceed $INT_{MAX}$ as long as the original input $x$ itself is within the $32$-bit range. This ingenious technique makes the solution both efficient and robust against overflow concerns.
