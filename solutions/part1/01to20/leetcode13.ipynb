```python
def romanToInt(s: str) -> int:
    roman_map = {
        'I': 1, 'V': 5, 'X': 10, 'L': 50,
        'C': 100, 'D': 500, 'M': 1000
    }
    total = 0
    n = len(s)
    i = 0
    
    while i < n:
        current_value = roman_map[s[i]]
        
        if i + 1 < n:
            next_value = roman_map[s[i+1]]
            
            if current_value < next_value:
                total += next_value - current_value
                i += 2
                continue
        
        total += current_value
        i += 1
            
    return total
```

## ðŸ›ï¸ LeetCode Problem 13: Roman to Integer - In Immense Detail

LeetCode problem 13, "Roman to Integer," is the reverse of LeetCode 12, requiring the conversion of a given Roman numeral string $S$ into its corresponding integer value. This problem tests the ability to parse a string, manage sequential reading, and correctly implement the rules of the Roman numeral system, particularly the crucial **subtractive principle**, using an iterative approach.

---

### ðŸ§ Understanding the Roman Numeral Structure

The Roman numeral system is generally additive, meaning symbols are summed from left to right (e.g., $\text{VI} = 5 + 1 = 6$). However, the system also uses a **subtractive principle** where a smaller value symbol immediately preceding a larger value symbol indicates subtraction. The six specific subtractive combinations are:
1.  $\text{IV}$ (4) and $\text{IX}$ (9)
2.  $\text{XL}$ (40) and $\text{XC}$ (90)
3.  $\text{CD}$ (400) and $\text{CM}$ (900)

The input string is guaranteed to be a valid Roman numeral representing a number between 1 and 3999, which simplifies the problem by removing the need for complex input validation.

---

### ðŸ’¡ The Right-to-Left (Reverse) Iteration Strategy

The most elegant and efficient strategy to handle the subtractive rule without complex lookaheads is to **iterate through the Roman numeral string from right to left** (from the least significant to the most significant position). This approach is based on the observation that when reading from right to left, if the current symbol's value is **less than** the value of the symbol immediately to its right (the previously processed symbol), it must be part of a subtractive pair and should be subtracted from the total sum instead of added.

---

### ðŸ’» Step-by-Step Algorithm Walkthrough

The algorithm begins by initializing a running `total` sum to $0$. It also maintains a variable, `prev_value`, which stores the integer value of the last Roman symbol processed.

1.  A mapping (e.g., a hash map or dictionary) is used to quickly convert each Roman character to its integer value ($\text{I} \to 1, \text{V} \to 5, \dots, \text{M} \to 1000$).
2.  The algorithm loops through the input string $S$ backward, from the last character to the first.
3.  In each iteration, the `current_value` of the current symbol $S[i]$ is retrieved from the map.
4.  The core logic compares the `current_value` with the `prev_value`:
    * **Additive Case (Current $\ge$ Previous):** If `current_value` is greater than or equal to `prev_value`, the current symbol is part of an additive sequence, and its value is **added** to the `total`. (This is the case for most symbols).
    * **Subtractive Case (Current $<$ Previous):** If `current_value` is strictly less than `prev_value`, the current symbol is the subtractive part (e.g., $\text{I}$ in $\text{IV}$). Its value is **subtracted** from the `total`.
5.  After the addition or subtraction, the `prev_value` is updated to the `current_value` to prepare for the next iteration.

### â±ï¸ Time Complexity and Efficiency

This right-to-left iteration is highly efficient and robust. Since the algorithm only makes a single pass over the string, and the work done inside the loop (map lookup, comparison, addition/subtraction) is $O(1)$ constant time, the overall time complexity is **$O(N)$**, where $N$ is the length of the Roman numeral string. Given the input constraint ($N$ is small, maximum length around 15 for 3999), the performance is excellent. The space complexity is $O(1)$ (or $O(N)$ if the input string is converted to a character array, but generally considered $O(1)$ for the fixed-size map).
