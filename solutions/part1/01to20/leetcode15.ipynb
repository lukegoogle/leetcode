```python
def threeSum(nums: list[int]) -> list[list[int]]:
    nums.sort()
    result = []
    n = len(nums)
    
    for i in range(n - 2):
        if i > 0 and nums[i] == nums[i - 1]:
            continue
            
        left, right = i + 1, n - 1
        
        while left < right:
            current_sum = nums[i] + nums[left] + nums[right]
            
            if current_sum == 0:
                result.append([nums[i], nums[left], nums[right]])
                
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                    
                left += 1
                right -= 1
                
            elif current_sum < 0:
                left += 1
            else:
                right -= 1
                
    return result
```

## ðŸŽ¯ LeetCode Problem 15: 3Sum - In Immense Detail

LeetCode problem 15, "3Sum," is a notoriously challenging array manipulation problem that requires finding all unique triplets $[a, b, c]$ in a given integer array $nums$ such that their sum equals zero ($a + b + c = 0$). This problem is the generalized version of the two-sum problem and is primarily focused on how to efficiently eliminate duplicate solutions and optimize the search from a cubic $O(N^3)$ approach down to a near-optimal quadratic $O(N^2)$.

---

### ðŸ§ Brute-Force and The Duplication Hurdle

A brute-force solution would involve using three nested loops to check every possible triplet $(i, j, k)$. This results in an immediate $O(N^3)$ time complexity, which is generally unacceptable for typical constraints. Furthermore, this method would generate numerous duplicate triplets. For example, if $nums = [-1, 0, 1, 2, -1, -4]$, the triplets $[-1, 0, 1]$ and $[1, 0, -1]$ represent the same set of numbers and should only be counted once. The challenge is therefore two-fold: optimize the time complexity and ensure the uniqueness of the returned triplets.

---

### ðŸ’¡ The $O(N^2)$ Optimized Strategy: Sorting and Two Pointers

The most effective strategy to solve the 3Sum problem in $O(N^2)$ time leverages two key techniques: **sorting** and the **Two-Pointer Technique**.

1.  **Sorting ($O(N \log N)$):** The array $nums$ is first sorted. Sorting is essential because it allows us to quickly skip duplicate numbers and enables the Two-Pointer Technique for efficient searching.
2.  **Fixing One Element ($O(N)$):** We iterate through the sorted array using a single pointer $i$. The element $nums[i]$ is fixed as the first element of our potential triplet. The target sum for the remaining two elements becomes $target' = 0 - nums[i]$.
3.  **Two Pointers for 2Sum ($O(N)$):** For each fixed $nums[i]$, the subproblem reduces to finding two numbers in the remaining subarray (from index $i+1$ to $N-1$) that sum up to $target'$. This is a classic 2Sum problem, which can be solved efficiently using two pointers, $L$ (starting at $i+1$) and $R$ (starting at $N-1$).

---

### âž¡ï¸ The Two-Pointer Movement and Triplet Handling

The two pointers, $L$ and $R$, move inward based on the sum $S = nums[L] + nums[R]$ compared to the required $target'$:
1.  **If $S < target'$:** The sum is too small. To increase the sum, we must include a larger number, so the left pointer moves right: $L++$.
2.  **If $S > target'$:** The sum is too large. To decrease the sum, we must include a smaller number, so the right pointer moves left: $R--$.
3.  **If $S = target'$:** A valid triplet $[nums[i], nums[L], nums[R]]$ has been found. This triplet is added to the result list.

---

### ðŸš« Duplication Elimination: The Crucial Step

Duplicate triplets must be carefully eliminated at two levels. This is where the sorting step becomes indispensable:

1.  **Skipping Duplicates for the Fixed Element ($nums[i]$):** After processing $nums[i]$, if the next element $nums[i+1]$ is the same as $nums[i]$, we skip $nums[i+1]$ by advancing $i$. This prevents processing the same first element multiple times.
2.  **Skipping Duplicates within the Two-Pointer Search ($nums[L]$ and $nums[R]$):** After a valid triplet is found, we must move $L$ and $R$ past any duplicates to ensure the next valid triplet is unique. We increment $L$ as long as $nums[L] = nums[L-1]$ and $L < R$. Similarly, we decrement $R$ as long as $nums[R] = nums[R+1]$ and $L < R$.

The overall time complexity is dominated by the initial sort ($O(N \log N)$) and the $N$ iterations of the outer loop, each containing a linear-time Two-Pointer search ($O(N)$). Thus, the total complexity is $O(N \log N + N^2)$, which simplifies to **$O(N^2)$**.
