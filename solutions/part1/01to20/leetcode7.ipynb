```python
def reverse(x: int) -> int:
    MAX_INT = 2**31 - 1
    MIN_INT = -2**31
    
    reversed_num = 0
    
    temp = abs(x)

    while temp != 0:
        digit = temp % 10
        temp //= 10
        
        # Positive overflow check
        if reversed_num > MAX_INT // 10 or (reversed_num == MAX_INT // 10 and digit > 7):
            return 0
        
        # Negative overflow check (magnitude check)
        # -2**31 is -2147483648 (last digit 8)
        if x < 0 and (reversed_num > MAX_INT // 10 or (reversed_num == MAX_INT // 10 and digit > 8)):
             return 0

        reversed_num = reversed_num * 10 + digit
        
    return reversed_num if x >= 0 else -reversed_num
```

LeetCode problem 7, "Reverse Integer," is a mathematical string manipulation problem that requires reversing the digits of a given 32-bit signed integer. While the core idea is simple, the primary challenge lies in handling the possibility of **integer overflow** during the reversal process.

### â“ The Reversal Mechanism

The task is to take an integer, $x$, and produce a new integer with its digits in the reverse order. For example, if $x = 123$, the output should be $321$; if $x = -123$, the output is $-321$; and if $x = 120$, the output is $21$. The standard technique for reversing digits involves a loop that repeatedly performs two operations: finding the last digit using the modulo operator ($x \pmod{10}$) and removing the last digit using integer division ($\lfloor x / 10 \rfloor$).

---

### âš ï¸ The Integer Overflow Constraint

The critical complexity of this problem stems from the requirement that the output must fit within the range of a standard 32-bit signed integer, which is $[ -2^{31}, 2^{31} - 1 ]$. If the reversed number exceeds this range, the function must return 0. The overflow check cannot wait until the final result is calculated; it must be performed **before** the final digit is appended to the reversed number in each iteration of the loop, as the addition or multiplication step itself can cause the overflow.

---

### ðŸ”Ž Pre-emptive Overflow Check

Let $rev$ be the reversed number so far, and $pop$ be the last digit popped from the original number $x$. In the next step, we want to update $rev$ to $rev_{new} = rev \times 10 + pop$. To check for potential overflow, we compare $rev$ against the maximum and minimum limits of a 32-bit integer, denoted as $INT_{MAX} = 2^{31} - 1$ and $INT_{MIN} = -2^{31}$.

For the positive case:
1.  If $rev > INT_{MAX} / 10$, then multiplying $rev$ by 10 will surely overflow.
2.  If $rev = INT_{MAX} / 10$, we check the last digit $pop$. If $pop > 7$ (since $INT_{MAX}$ ends in 7), the addition will cause an overflow.

For the negative case (using the absolute values for comparison):
1.  If $rev < INT_{MIN} / 10$, then multiplying $rev$ by 10 will surely underflow.
2.  If $rev = INT_{MIN} / 10$, we check the last digit $pop$. If $pop < -8$ (since $INT_{MIN}$ ends in 8), the addition will cause an underflow.  If any of these conditions are met, the loop must terminate and return 0 immediately.

---

### ðŸ’» The Final Algorithm

The algorithm proceeds by setting the `rev` variable to 0. It iterates while $x$ is not zero, extracting the $pop$ digit. Before updating $rev = rev \times 10 + pop$, the pre-emptive overflow check is performed. If no overflow is detected, $rev$ is updated, and $x$ is truncated using integer division. Once $x$ becomes 0, the loop terminates, and the correctly reversed integer $rev$ is returned. This method correctly handles positive and negative numbers and ensures the result adheres to the 32-bit integer constraints.
