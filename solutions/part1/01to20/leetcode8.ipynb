```python
def myAtoi(s: str) -> int:
    i = 0
    n = len(s)
    
    MAX_INT = 2**31 - 1
    MIN_INT = -2**31
    
    while i < n and s[i] == ' ':
        i += 1
        
    sign = 1
    if i < n and (s[i] == '+' or s[i] == '-'):
        if s[i] == '-':
            sign = -1
        i += 1
        
    result = 0
    while i < n and s[i].isdigit():
        digit = int(s[i])
        
        # Check against the positive limit (abs value check)
        if result > MAX_INT // 10 or (result == MAX_INT // 10 and digit > 7):
            return MAX_INT if sign == 1 else MIN_INT
            
        result = result * 10 + digit
        i += 1
        
    return sign * result
```

### ðŸ“ The Four Conversion Rules

The conversion process must follow a strict sequence of four steps, corresponding to how characters are read and interpreted from the input string $S$:

1.  **Whitespace Skipping:** The algorithm must first discard all leading whitespace characters (only spaces ' ') until the first non-whitespace character is found.
2.  **Sign Detection:** The next character must be checked. If it is a '-' or a '+', it is recorded as the sign of the result, and the reading position moves to the next character. Only the first sign encountered is valid; multiple signs are treated as an invalid sequence.
3.  **Digit Extraction:** The algorithm must then read all subsequent characters until a non-digit character is encountered or the end of the string is reached. These characters form the numerical part of the result. If the first non-whitespace character is not a sign or a digit, no valid conversion can be performed.
4.  **Final Clamping (Overflow/Underflow):** The numerical value extracted is then converted to an integer. Before the final return, this integer must be clamped (or saturated) to fit within the 32-bit signed integer range, $[-2^{31}, 2^{31} - 1]$.

### ðŸš« Handling Invalid Input and Initial State

A common point of failure is when the string doesn't conform to the expected pattern after whitespace stripping. If the first non-whitespace character is neither a digit ('0'-'9') nor a sign ('+' or '-'), the conversion process must be stopped immediately, and the function must return $0$. Similarly, if the string only contains whitespace, or if it is empty after trimming, the result is $0$. The algorithm requires careful state tracking to ensure that a sign is only accepted once and only as the first non-whitespace, non-sign character.

### âš ï¸ The Critical Overflow/Underflow Detection During Digit Accumulation

As digits are parsed and accumulated into the `result` integer, the possibility of exceeding $INT_{MAX}$ ($2,147,483,647$) or falling below $INT_{MIN}$ ($-2,147,483,648$) arises. Similar to LeetCode 7, the check must be **pre-emptive** in the accumulation loop before the operation $result_{new} = result \times 10 + digit$ is executed. Let $INT_{MAX\_DIV\_10} = INT_{MAX} / 10$ and $INT_{MIN\_DIV\_10} = INT_{MIN} / 10$.

For positive numbers:
1.  If $result > INT_{MAX\_DIV\_10}$, multiplication by 10 guarantees overflow.
2.  If $result = INT_{MAX\_DIV\_10}$ and the next digit $d$ is greater than $7$ (since $INT_{MAX}$ ends in $7$), overflow is guaranteed.

For negative numbers (when accumulating as a negative value):
1.  If $result < INT_{MIN\_DIV\_10}$, multiplication by 10 guarantees underflow.
2.  If $result = INT_{MIN\_DIV\_10}$ and the next digit $d$ is greater than $8$ (since $INT_{MIN}$'s absolute value ends in $8$), underflow is guaranteed.

If an overflow or underflow is detected at any point during digit accumulation, the loop should immediately break, and the clamped value ($INT_{MAX}$ or $INT_{MIN}$) should be returned.

Once the digit extraction loop terminates (due to a non-digit character, the end of the string, or an overflow detection), the final `result` is generated.
